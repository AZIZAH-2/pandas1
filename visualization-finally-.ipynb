```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('diabetes.csv')
df.head()
```

```python
df.columns
```

No missing values

```python
print(df['Pregnancies'].isnull().values.any())
print(df['BloodPressure'].isnull().values.any())
print(df['Glucose'].isnull().values.any())
print(df['Age'].isnull().values.any())
print(df['BMI'].isnull().values.any())
```

**We take the youth age group**

```python
df = df[df['Age'] < 50]
```

```python
plt.figure(figsize=[10,8])
df['Insulin'].value_counts().sort_index().plot.line()
```

```python
plt.figure(figsize=[20,10])
order = list(df['Pregnancies'].value_counts().index)
base_color = sns.color_palette()[3]
a=df.Age
tv_g = sns.countplot(data=df,x='Pregnancies',hue=a, order=order, color=base_color)
tv_g.set_xlabel('',fontsize = 15)
tv_g.set_ylabel('',fontsize = 15)
tv_g.set_title("",fontsize = 20)
plt.legend(title = '',)
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='Outcome', data=df, palette='viridis')
```

```python
plt.figure(figsize=(10,5))
sns.distplot(df['Age'])
```

```python
plt.figure(figsize=[20,15])
data = df['Age'].value_counts()
data.plot(kind='pie', autopct='%0.1f%%')
```
